```python
color_map = {v: list(mcolors.CSS4_COLORS.keys())[i+5] for i, v in enumerate(df["neighbourhood"].unique())}
color_map
```

```python
# Geographic map of 10,000 listings in Athens, Greece.


# Set up the map centered around Athens
athens_map = folium.Map([37.9838, 23.7275], zoom_start=11.5, tiles='CartoDB dark_matter')

# Filter the data
total_points = 10000
filtered_df = df.head(total_points)
filtered_df['name'] = filtered_df['name'].fillna('')

# Add listings to the map
for idx, row in filtered_df.iterrows():
    location = [row['latitude'], row['longitude']]
    name = row['name']
    neighborhood_color = color_map[row['neighbourhood']]
    folium.CircleMarker(
        location=location,
        popup=name,
        color=neighborhood_color,
        radius=1
    ).add_to(athens_map)

print("Total points: ", len(filtered_df))

athens_map.save('athens_listing_map.html')
```

```python
# Geographic map of 30,000 listings grouped by neighborhood.




# Define the map centered at Athens
athens_map = folium.Map([37.9838, 23.7275], zoom_start=11.5, tiles='CartoDB dark_matter')

# Select the first 10000 rows of the DataFrame
df_sample = df.head(10000)

# Extract the name series
name_series = df_sample['name'].fillna("").reset_index(drop=True)
color = [color_map[i] for i in df_sample["neighbourhood"]]

# Add the markers to the map
for idx, (lat, long) in enumerate(df_sample[["latitude", "longitude"]].values):
    folium.CircleMarker([lat, long], 
                        popup=name_series[idx], 
                        color=color[idx], 
                        radius=1).add_to(athens_map)

# Print the total number of points
print("Total points:", len(df_sample))

# Define the template for the draggable legend
template1 = """
{% macro html(this, kwargs) %}
<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>jQuery UI Draggable - Default functionality</title>
  <link rel="stylesheet" href="//code.jquery.com/ui/1.12.1/themes/base/jquery-ui.css">
  <script src="https://code.jquery.com/jquery-1.12.4.js"></script>
  <script src="https://code.jquery.com/ui/1.12.1/jquery-ui.js"></script>
  <script>
  $( function() {
    $( "#maplegend" ).draggable({
                    start: function (event, ui) {
                        $(this).css({
                            right: "auto",
                            top: "auto",
                            bottom: "auto"
                        });
                    }
                });
});
  </script>
</head>
<body>
<div id='maplegend' class='maplegend' 
     style='position: absolute; z-index:9999; border:2px solid grey; background-color:rgba(255, 255, 255, 0.8);
     border-radius:6px; padding: 10px; font-size:14px; right: 10px; bottom: 40px;'>
...
<div class='legend-title'>Legend</div>
<div class='legend-scale'>
  <ul class='legend-labels'>
"""

template2 = """
  </ul>
</div>
</div>
</body>
</html>
<style type='text/css'>
  .maplegend .legend-title {
    text-align: left;
    margin-bottom: 5px;
    font-weight: bold;
    font-size: 90%;
    }
  .maplegend .legend-scale ul {
    margin: 0;
    margin-bottom: 5px;
    padding: 0;
    float: left;
    list-style: none;
    }
  .maplegend .legend-scale ul li {
    font-size: 80%;
    list-style: none;
    margin-left: 0;
    line-height: 18px;
    margin-bottom: 2px;
    }
  .maplegend ul.legend-labels li span {
    display: block;
    float: left;
    height: 16px;
    width: 30px;
    margin-right: 5px;
    margin-left: 0;
    border: 1px solid #999;
    }
  .maplegend .legend-source {
    font-size: 80%;
    color: #777;
    clear: both;
    }
  .maplegend a {
    color: #777;
    }
</style>
{% endmacro %}
"""

legend_items = [f"<li><span style='background:{color};opacity:0.7;'></span>{neigh}</li>" 
                for neigh, color in color_map.items()]

legend = "\n".join(legend_items)
template = template1 + legend + template2
macro = MacroElement()
macro._template = Template(template)

# Add the legend to the map
athens_map.get_root().add_child(macro)
athens_map.save('athens_neighbourhood_map.html')

```

```python
# Geographic map of Athens, Greece, categorized by room type.


# Define the color map for room types
room_type_colors = {
    'Entire home/apt': 'red',
    'Private room': 'green',
    'Hotel room': 'orange',
    'Shared room': 'blue',
}

# Create the map centered at Athens
athens_map = folium.Map([37.9838, 23.7275], zoom_start=11.5, tiles='CartoDB dark_matter')

# Select the first 30000 rows of the DataFrame
samples = 10000
df_sample = df.head(samples)

# Extract the name series and room type
name_series = df_sample['name'].fillna("").reset_index(drop=True)
room_type = df_sample['room_type'].reset_index(drop=True)

# Add the markers to the map
for idx, (lat, long) in enumerate(df_sample[["latitude", "longitude"]].values):
    r_type = room_type[idx]
    color = room_type_colors.get(r_type, "grey")
    popup_content = f"{name_series[idx]}<br>Room Type: {r_type}"
    folium.CircleMarker([lat, long], 
                        popup=popup_content,
                        color=color,
                        fill=True,
                        fill_opacity=0.7,
                        radius=1).add_to(athens_map)

# Print the total number of points
print("Total points: ", len(df_sample))



legend = """
    <li><span style='background:red;opacity:0.7;'></span>Entire home/apt</li>
    <li><span style='background:orange;opacity:0.7;'></span>Hotel room</li>
    <li><span style='background:green;opacity:0.7;'></span>Private room</li>
    <li><span style='background:blue;opacity:0.7;'></span>Shared room</li>
"""

template = template1+legend+template2
macro = MacroElement()
macro._template = Template(template)

athens_map.get_root().add_child(macro)
athens_map.save('mapRoomType.html')
athens_map


# Display the map
athens_map.save('athens_roomtype_map.html')
```
