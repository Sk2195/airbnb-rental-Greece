```python
# Encode categorical variables

def one_hot_encode_column(df, column_name):
    """One-hot encodes a specified column of a dataframe."""
    # Label encoding
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(df[column_name].fillna(""))

    # OneHot encoding
    onehot_encoder = OneHotEncoder(sparse=False)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded.reshape(-1, 1))

    # Constructing column names for one-hot encoded columns
    onehot_cols = [f"{column_name}_{cls}" for cls in label_encoder.classes_]
    
    return pd.DataFrame(onehot_encoded, columns=onehot_cols)

df_cat = df[['neighbourhood', 'room_type', 'property_type', 'NearestStation', 'IsHoliday', 'nearestHoliday']]

# Using the function to one-hot encode each column
for cols in df_cat.columns:
    df_encoded = one_hot_encode_column(df_cat, cols)
    df_cat = pd.concat([df_cat, df_encoded], axis=1)
    df_cat.drop(cols, axis=1, inplace=True)

# Adding the 'price' column after the loop
df_cat = pd.concat([df['price'], df_cat], axis=1)
```
